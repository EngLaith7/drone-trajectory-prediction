```python
from pathlib import Path
import pandas as pd

# ========================
#  Load Data
# ========================

#make sure that you copied the imu_data.csv file to the data folder
data_path = Path.cwd().parent/ "data" / "imu_data.csv"

if not data_path.exists():
    raise FileNotFoundError(f"‚ö†Ô∏è The file is not found at :{data_path}")

df = pd.read_csv(data_path)

# ========================
# 1. Basic Information
# ========================
print("üîπ Data Info:")
print(df.info())

print("\nüîπ Missing Values:")
print(df.isnull().sum())

print("\nüîπ Descriptive Statistics:")
print(df.describe().T)

```


 Distribution of values (digitally)

```python
# ========================
# Distribution of values (digitally)
# ========================
print("\nüîπ Accelerometer stats:")
print(df[['accel_x','accel_y','accel_z']].describe().T)

print("\nüîπ Gyroscope stats:")
print(df[['gyro_x','gyro_y','gyro_z']].describe().T)

print("\nüîπ Magnetometer stats:")
print(df[['mag_x','mag_y','mag_z']].describe().T)

print("\nüîπ Position stats:")
print(df[['pos_x','pos_y','pos_z']].describe().T)

print("\nüîπ Orientation stats (Roll, Pitch, Yaw):")
print(df[['roll','pitch','yaw']].describe().T)
```
